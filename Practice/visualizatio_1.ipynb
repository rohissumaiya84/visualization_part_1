## import library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

Line plot

```python
x=[10,15,20,25,30,35,40]
y=[100,150,200,250,300,350,400]

plt.figure(figsize=(4,3))
plt.title("Line Plot")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
 
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, "o", color="red")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
plt.title("Line Plot")
 
plt.show()
```

## Scattar plot - Matlablib

```python
x=[10,15,20,25,30,35,40]
y=[100,150,300,250,300,80,400]

plt.figure(figsize=(4,3))
plt.title("Scatter Plot")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
plt.scatter(x,y, color="g")
plt.show()
```

```python
z = ["red", "blue","red", "blue","red", "blue","red"]
plt.figure(figsize=(4,3))
plt.title("Scatter Plot")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
plt.scatter(x,y, color=z, alpha=0.5, s=100)
plt.show()
```

## Scattar plot -Seaborn

```python
z = ["apple", "berry","strawberry", "orange","mango", "jack","banana"]
plt.figure(figsize=(4,3))
plt.xlabel("value of x")
plt.ylabel("value of y")
plt.title("fruit value identificartion")
sns.scatterplot(x=x,y=y, hue =z, s=100)
plt.show()
```

# bar plot - matplotlib

```python
col = ["red", "blue","red", "blue","red", "blue","red"]
plt.figure(figsize=(5,4))
plt.title("Scatter Plot")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
plt.bar(x,y, color=col, alpha=0.6, width=2, edgecolor="black", linewidth=2 )
plt.show()
```

# bar plot - seaborn

```python
z = ["apple", "berry","strawberry", "orange","mango", "jack","banana"]
plt.figure(figsize=(6.5,4))
plt.xlabel("value of x")
plt.ylabel("value of y")
plt.title("fruit value identificartion")
sns.barplot(x=z,y=y, hue =z,width=0.6,edgecolor="black", linewidth=2),
plt.show()
```

```python
x=[10,15,20,25,30,35,40]
y=[100,150,300,250,300,80,400]
col = ["red", "blue","red", "blue","red", "blue","red"]
z = ["apple", "berry","strawberry", "orange","mango", "jack","banana"]

plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
 
plt.title("Scatter Plot")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
plt.scatter(x,y, color="g")

plt.subplot(1,3,2)
 
plt.title("Line Plot")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
plt.plot(x,y, "--", color="r")


plt.subplot(1,3,3)
 
plt.title("Bar Plot")
plt.xlabel("Values of X")
plt.ylabel("Values of Y")
plt.bar(x,y, color=col, alpha=0.6, width=2, edgecolor="black", linewidth=2 )

plt.show()
```
